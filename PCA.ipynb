## Principal Component Analysis

```python
## Importing the required libraries

import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
%matplotlib inline

from sklearn.datasets import load_breast_cancer
```

```python
# Cancer dataset from sci-kit learn

cancer = load_breast_cancer()
```

```python
# Checking keys

cancer.keys()
```

```python
# Checking the target key values

print(cancer["target"])
```

```python
#Converting data into a dataframe

df = pd.DataFrame(cancer["data"], columns=cancer["feature_names"])
```

```python
df.head()
```

```python
df.describe()
```

```python
df.shape
```

```python
# Scaling will scale the values such that the distribution is now centred around 0, with a standard deviation of 1

from sklearn.preprocessing import StandardScaler

scaler=StandardScaler()
scaler.fit(df)
```

```python
scaled_data = scaler.transform(df)
```

```python
## Checking the values in the scaled data

scaled_data
```

```python
## Importing module for PCA 

from sklearn.decomposition import PCA
```

```python
pca = PCA(n_components = 2) ## How many dimensions, we want to reduce?
```

```python
pca.fit(scaled_data) ## Fitting the scaled data
```

```python
x_pca = pca.transform(scaled_data)
```

```python
scaled_data.shape
```

```python
x_pca.shape
```

```python
plt.figure(figsize=(8,6))

plt.scatter(x_pca[:,0],x_pca[:,1],c=cancer['target'])

plt.xlabel('First principle component')
plt.ylabel('Second principle component')

plt.show()
```
